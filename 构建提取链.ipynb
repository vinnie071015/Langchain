```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

# å®šä¹‰ä¸€ä¸ªä¸ªæ€§åŒ–çš„æç¤ºè¯æ¥æä¾›æŒ‡ç¤ºå’Œä»»ä½•é¢å¤–çš„ä¸Šä¸‹æ–‡ã€‚
# 1) ä½ å¯ä»¥æ·»åŠ æ ·ä¾‹åˆ°æç¤ºè¯æ¨¡ç‰ˆä¸­æ¥ä¼˜åŒ–æå–è´¨é‡ã€‚
# 2) å¼•å…¥é¢å¤–çš„å‚æ•°æŠŠä¸Šä¸‹æ–‡è€ƒè™‘è¿›å»ã€‚ï¼ˆä¾‹å¦‚æ–‡æ¡£çš„Metadataå±æ€§ï¼‰


prompt_template = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are an expert extraction algorithm."
            "Only extract relevant information from the text."
            "If you do not know the value of an attribute asked to extract,"
            "return null for the attribute's value.",
        ),
        # è¯·æŸ¥çœ‹How-toæ–‡æ¡£æ¥ä¼˜åŒ–å‚è€ƒæ ·ä¾‹çš„æ€§èƒ½
        ("human","{text}"),
    ]
)
```

- æˆ‘ä»¬éœ€è¦ä½¿ç”¨æ”¯æŒfunctionæˆ–è€…tool callingçš„å¤§æ¨¡å‹

```python
import getpass
import os 
# å‰ç½®éœ€è¦è¿›å…¥deepseekå®˜ç½‘æ³¨å†Œå¹¶ç”ŸæˆAPI_KEYï¼Œè®°å¾—ä¿å­˜å¥½ï¼Œå¦åˆ™ä¸¢å¤±åéœ€è¦é‡æ–°ç”Ÿæˆã€‚
# è¿™ä¸€è¡Œæ£€æŸ¥ç¨‹åºæ˜¯å¦å·²ç»æœ‰ Deepseek API å¯†é’¥ã€‚å¦‚æœæ²¡æœ‰ï¼Œå°±éœ€è¦ç”¨æˆ·æ‰‹åŠ¨è¾“å…¥ã€‚
if not os.getenv("DEEPSEEK_API_KEY"):
    os.environ["DEEPSEEK_API_KEY"] = getpass.getpass("Enter your DeepSeek API key: ")

# é…ç½®æ¥å£å‚æ•°ï¼Œåˆ›å»ºè¯­è¨€æ¨¡å‹çš„å®ä¾‹
from langchain_deepseek import ChatDeepSeek

llm = ChatDeepSeek(
    model="deepseek-chat", #æŒ‡å®šä½¿ç”¨çš„æ¨¡å‹
    temperature=0, #æ§åˆ¶ç”Ÿæˆæ–‡æœ¬çš„éšæœºæ€§
    max_tokens=None, #ç”Ÿæˆçš„æœ€å¤§tokenæ•°é™åˆ¶
    timeout=None, #è¯·æ±‚è¶…æ—¶æ—¶é—´
    max_retries=2, #è¯·æ±‚å¤±è´¥æ—¶çš„æœ€å¤§é‡è¯•æ¬¡æ•°
    # other params...
)
```

```python
# æ„å»ºæ¨¡å‹æ•°æ®è¾“å‡ºå®ä¾‹ï¼ŒæŒ‰ç…§Personçš„æ•°æ®ç»“æ„è¾“å‡ºç»“æœæ–‡æœ¬
structured_llm = llm.with_structured_output(schema=Person)
```

```python
from typing import List, Optional 

from pydantic import BaseModel, Field 

class Person(BaseModel):
    """Information about a person."""
    
    # å®ä½“Personçš„æ–‡æ¡£å­—ç¬¦ä¸²
    # è¿™ä¸ªæ–‡æ¡£å­—ç¬¦ä¸²ä¼šè¢«å‘ç»™å¤§æ¨¡å‹ä½œä¸ºPersonæ•°æ®è¾“å‡ºç»“æ„çš„æè¿°ï¼Œè¿™èƒ½å¤Ÿå¸®åŠ©ä¼˜åŒ–æ•°æ®æå–ç»“æœ
    
    # æç¤ºï¼š
    # 1. æ¯ä¸ªå­—æ®µéƒ½æ˜¯å¯é€‰çš„ -- è¿™å…è®¸å¤§æ¨¡å‹ä¸æå–è¿™ä¸ªå­—æ®µï¼Œåœ¨æŸäº›ä¿¡æ¯ä¸å­˜åœ¨çš„æƒ…å†µä¸‹
    # 2. æ¯ä¸ªå­—æ®µéƒ½æœ‰æè¿° -- è¿™ä¸ªæè¿°ä¼šè¢«å¤§æ¨¡å‹é‡‡ç”¨
    # å¥½çš„æè¿°èƒ½å¤Ÿå¸®åŠ©ä¼˜åŒ–æå–ç»“æœ
    
    name: Optional[str] = Field(default=None, description="The name of the person")
    hair_color: Optional[str] = Field(
        default=None, description="The color of the person's hair if known"
    )
    height_in_meters: Optional[str] = Field(
        default=None, description="Height measured in meters"
    )
    
class Data(BaseModel):
    """Extract data about people."""
    
    # Creates a model so that we can extract multiple entities.
    people: List[Person]
```

```python
from langchain_core.utils.function_calling import tool_example_to_messages

examples = [
    (
        "The ocean is vast and blue. It's more than 20,000 feet deep.",
        Data(people=[]),
    ),
    (
        "Fiona traveled far from France to Spain.",
        Data(people=[Person(name="Fiona", height_in_meters=None, hair_color=None)]),
    ),

]

messages = []

for txt, tool_call in examples:
    if tool_call.people:
        # å¯¹äºä¸€éƒ¨åˆ†å¤§æ¨¡å‹ï¼Œæœ€ç»ˆçš„ç»“æœæ˜¯å¯é€‰çš„
        ai_response = "Detected people."
    else:
        ai_response = "Detected no people."
    messages.extend(tool_example_to_messages(txt, [tool_call], ai_response=ai_response))
    
    
```

```python
for message in messages:
    message.pretty_print()
```
